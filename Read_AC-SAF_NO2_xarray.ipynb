```python
import xarray as xr
import cartopy.crs as ccrs
from matplotlib import pyplot as plt

data=xr.open_dataset("GOME_NO2_Global_201305_METOPB_DLR_v1.nc")
prod=xr.open_dataset("GOME_NO2_Global_201305_METOPB_DLR_v1.nc", group="PRODUCT")

# Should read it again and associate the array to the data

data["NO2total"]=prod.NO2total

plt_data=data.sel({"latitude": slice(30,70),"longitude": slice(-20,30)})

%matplotlib inline
plt_data.NO2total.plot()
```

```python
fig=plt.figure
#fig.set_dpi(400)
spl=plt.axes(projection=ccrs.PlateCarree())
plt.contour(plt_data.longitude,plt_data.latitude,plt_data.NO2total,100,transform=ccrs.PlateCarree())
spl.coastlines()
spl.gridlines()
plt.colorbar()
```
